## Data Preparation and Cleaning

As the dataset contains large amount of data. It is needed to see if their is any missing value in it before working on it so for that i will be using different libraries for further processing, preparation and cleaning of data.



Before working further lets import all the libraries that i am going to use during analysing.

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

Lets first read the data by using read_csv function and convert this data into a dataframe.

```python
reliance_df = pd.read_csv('RELIANCE_2020.csv')
reliance_df
```

```python
#Lets check no. of rows and col in dataset

reliance_df.shape
```

So it seems their are 15 col and 252 rows in reliance dataframe.

Now lets check the basic information about the data.

```python
#info() function gives overall information about the data.

reliance_df.info()
```

```python
reliance_df.Date
```

The Date column has datatype object lets change it to datetime datatype.

```python
reliance_df['Date'] = pd.to_datetime(reliance_df.Date)
```

```python
reliance_df.Date
```

The date column datatype has changed now its easy to access it. lets make different parts of Date column in different columns.



```python
reliance_df['month'] = pd.DatetimeIndex(reliance_df.Date).month
reliance_df['day'] = pd.DatetimeIndex(reliance_df.Date).day
reliance_df['weekday'] = pd.DatetimeIndex(reliance_df.Date).weekday
reliance_df
```

We have added month, day, weekday columns in dataframe.

```python
# columns function gives all the columns in the data.

reliance_df.columns
```

```python
#lets just check some top 10 values

reliance_df.head(10)
```

From above we dont require some columns so lets just delete them. We dont require symbol and series column so lets just delete them.

```python
reliance_df.drop(['Symbol', 'Series'], axis=1)
```

```python
reliance_df.describe()
```

## Exploratory Analysis and Visualization

In this, we'll do some visualization using some plots, charts in order to make some exploratory analysis.



Let's begin by importing`matplotlib.pyplot` and `seaborn`.

```python
import seaborn as sns
import matplotlib
import matplotlib.pyplot as plt
%matplotlib inline

sns.set_style('darkgrid')
matplotlib.rcParams['font.size'] = 14
matplotlib.rcParams['figure.figsize'] = (9, 5)
matplotlib.rcParams['figure.facecolor'] = '#00000000'
```

### 1. What was the maximum highest value of Reliance stock in the year 2020 ?

```python
reliance_df_high = reliance_df.sort_values(by= 'High', ascending=False).head(10)
reliance_df_high
```

```python
sns.barplot(reliance_df.month, reliance_df.High);
```

It seems that in the month of september and october the stocks value was high which was 2369.

### 2. What was the lowest close in the year 2020 of the stocks ?

```python
reliance_df_close = reliance_df.sort_values(by='Prev Close', ascending=True).head(10)
reliance_df_close
```

```python
sns.histplot(reliance_df['Prev Close']);
```

### 3. What was the turnover in the month of september ?

```python
# lets extract september month data
reliance_df_september = reliance_df[reliance_df.month == 9]

# data of september month turnover
reliance_df_turnoversept = reliance_df_september['Turnover']

# sum of turnover in september
reliance_df_turntotal = reliance_df_turnoversept.sum()
reliance_df_turntotal
```

```python
plt.title('Turnover in September')
plt.plot(reliance_df_turnoversept);
```

### 4.  What was the count of trade happened in Reliance Shares 2020 ?

```python
reliance_df_trade = reliance_df.Trades
reliance_df_tradecount = reliance_df_trade.sum()
reliance_df_tradecount
```

```python
sns.barplot(reliance_df.month, reliance_df.Trades);
```

It seems the trade during the month of april, july were high as compared to other months.

### 5. What was the volume of trades in each month ? 

```python
plt.title('Volume of Trades in Each month')
sns.barplot(reliance_df.month, reliance_df.Volume);
```

### 6. What was the highest opening value of Reliance share ?

```python
reliance_df_open = reliance_df.sort_values(by='Open', ascending=False).head(10)
reliance_df_open
```

So the highest opening value was 2319.75 which was in the month of September.
